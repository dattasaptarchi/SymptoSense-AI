## Set up the database

### Subtask:
Define the database schema to store patient information, doctor availability, appointment schedules, and potentially historical data for scheduling optimization. Choose a suitable database (e.g., SQLite for simplicity in this context).


**Reasoning**:
The subtask requires defining a database schema for patient information, doctor availability, and appointment schedules using SQLite. This involves importing the necessary library, establishing a connection, and creating the specified tables with appropriate columns and constraints.



```python
import sqlite3

# Establish connection to the database
conn = sqlite3.connect('symptosense_ai.db')
cursor = conn.cursor()

# Create patients table
cursor.execute('''
CREATE TABLE IF NOT EXISTS patients (
    patient_id INTEGER PRIMARY KEY,
    name TEXT,
    symptoms TEXT,
    urgency_level TEXT
)
''')

# Create doctors table
cursor.execute('''
CREATE TABLE IF NOT EXISTS doctors (
    doctor_id INTEGER PRIMARY KEY,
    name TEXT,
    specialty TEXT,
    room_number TEXT  -- Added room number column
)
''')

# Create availability table
cursor.execute('''
CREATE TABLE IF NOT EXISTS availability (
    availability_id INTEGER PRIMARY KEY,
    doctor_id INTEGER,
    day TEXT,
    time_slot TEXT,
    FOREIGN KEY (doctor_id) REFERENCES doctors (doctor_id)
)
''')

# Create appointments table
cursor.execute('''
CREATE TABLE IF NOT EXISTS appointments (
    appointment_id INTEGER PRIMARY KEY,
    patient_id INTEGER,
    doctor_id INTEGER,
    day TEXT,
    time_slot TEXT,
    status TEXT,
    FOREIGN KEY (patient_id) REFERENCES patients (patient_id),
    FOREIGN KEY (doctor_id) REFERENCES doctors (doctor_id)
)
''')

# Commit changes and close the connection
conn.commit()
conn.close()
```

## Implement scheduling logic

### Subtask:
Create an AI agent (using rule-based logic or simple algorithms, as APIs are not allowed) to schedule appointments based on triage results, doctor availability, and existing appointments.


**Reasoning**:
Define the `schedule_appointment` function to handle appointment scheduling based on urgency and availability, retrieving and updating data in the database.



```python
import sqlite3
import random
from datetime import datetime, timedelta

def schedule_appointment(patient_id, urgency_level, suggested_department):
    conn = sqlite3.connect('symptosense_ai.db')
    cursor = conn.cursor()

    # Determine the timeframe based on urgency level
    if urgency_level == "high":
        timeframe_days = 1 # Schedule within 1 day
    elif urgency_level == "medium":
        timeframe_days = 3 # Schedule within 3 days
    else: # low urgency
        timeframe_days = 7 # Schedule within 7 days

    # Get available doctors for the suggested department
    cursor.execute("SELECT doctor_id FROM doctors WHERE specialty = ?", (suggested_department,))
    available_doctors = [row[0] for row in cursor.fetchall()]

    if not available_doctors:
        conn.close()
        return {"status": "failure", "message": f"No doctors available for {suggested_department}."}

    # Find available time slots within the timeframe, prioritizing based on urgency
    today = datetime.now()
    for i in range(timeframe_days + 1):
        target_date = today + timedelta(days=i)
        target_day = target_date.strftime('%Y-%m-%d')

        # Shuffle the list of available doctors to distribute appointments
        random.shuffle(available_doctors)

        for doctor_id in available_doctors:
            # Get availability for the doctor on the target day
            cursor.execute("SELECT time_slot FROM availability WHERE doctor_id = ? AND day = ?", (doctor_id, target_day))
            available_slots = [row[0] for row in cursor.fetchall()]

            # Check for existing appointments for the doctor on the target day
            cursor.execute("SELECT time_slot FROM appointments WHERE doctor_id = ? AND day = ?", (doctor_id, target_day))
            booked_slots = [row[0] for row in cursor.fetchall()]

            # Find available slots that are not booked
            possible_slots = [slot for slot in available_slots if slot not in booked_slots]

            if possible_slots:
                # Select a random available slot (simple approach)
                selected_time_slot = random.choice(possible_slots)

                # Insert the new appointment
                cursor.execute("INSERT INTO appointments (patient_id, doctor_id, day, time_slot, status) VALUES (?, ?, ?, ?, ?)",
                               (patient_id, doctor_id, target_day, selected_time_slot, "Scheduled"))

                conn.commit()
                conn.close()
                return {"status": "success", "message": "Appointment scheduled successfully.",
                        "appointment_details": {"patient_id": patient_id, "doctor_id": doctor_id,
                                                "day": target_day, "time_slot": selected_time_slot}}

    conn.close()
    return {"status": "failure", "message": "Could not find an available appointment within the timeframe."}

# Example usage (optional, for testing)
# Assume a patient with patient_id = 1, urgency_level = "high", suggested_department = "cardiology"
# Note: You would need to have doctors and availability data in the database for this to work.
# result = schedule_appointment(1, "high", "cardiology")
# print(result)
```

## Integrate components

### Subtask:
Connect the database, triage, scheduling, and user input handling components to create a unified system.


**Reasoning**:
Create the main function to orchestrate the workflow, including getting user input, connecting to the database, inserting patient data, analyzing symptoms, scheduling the appointment, and printing the result, finally closing the database connection.



```python
import sqlite3

def main():
    """Orchestrates the workflow for the SymptoSense AI system."""
    print("Welcome to SymptoSense AI Patient Scheduling and Triage System")

    # 1. Get patient input
    name, symptoms = get_patient_input()

    conn = None  # Initialize connection to None
    try:
        # 2. Establish database connection
        conn = sqlite3.connect('symptosense_ai.db')
        cursor = conn.cursor()

        # 3. Insert patient data and get patient_id
        cursor.execute("INSERT INTO patients (name, symptoms) VALUES (?, ?)", (name, symptoms))
        patient_id = cursor.lastrowid
        conn.commit()
        print(f"Patient '{name}' added with ID: {patient_id}")

        # 4. Analyze symptoms
        urgency_level, suggested_department = analyze_symptoms(symptoms)
        print(f"Symptom Analysis: Urgency Level - {urgency_level}, Suggested Department - {suggested_department}")

        # Update patient record with urgency level (optional but good practice)
        cursor.execute("UPDATE patients SET urgency_level = ? WHERE patient_id = ?", (urgency_level, patient_id))
        conn.commit()

        # 5. Schedule appointment
        appointment_result = schedule_appointment(patient_id, urgency_level, suggested_department)

        # 6. Print result to user
        if appointment_result["status"] == "success":
            details = appointment_result["appointment_details"]
            print("\nAppointment Scheduled Successfully!")
            print(f"Appointment Details:")
            print(f"  Patient ID: {details['patient_id']}")
            print(f"  Doctor ID: {details['doctor_id']}")
            print(f"  Date: {details['day']}")
            print(f"  Time: {details['time_slot']}")
        else:
            print(f"\nFailed to schedule appointment: {appointment_result['message']}")

    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if conn:
            conn.rollback() # Rollback changes if something goes wrong
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        # 7. Close database connection
        if conn:
            conn.close()
            print("Database connection closed.")

# Assuming get_patient_input, analyze_symptoms, and schedule_appointment are defined in the environment
# You would typically import them from their respective modules

# Example of how you might call the main function if this were a script:
# if __name__ == "__main__":
#     main()

# For this notebook environment, we will call main directly for demonstration
main()
```
